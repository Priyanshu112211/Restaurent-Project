## Import Important Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')
```

## Reading  csv File

```python
df1=pd.read_csv('zomato.csv')
```

```python
df1.head()
```

## Find Shape and columns

```python
df1.shape
```

```python
df1.columns
```

## Drop Unnecessary Columns

```python
df1=df1.drop(['url','address','phone','reviews_list', 'menu_item', 'dish_liked'], axis=1)
df1.head()
```

```python
df1.info()
```

## Drop Duplicates

```python
df1.drop_duplicates(inplace=True)
```

```python
df1.shape
```

## Finding Unique values in rate column

```python
df1['rate'].unique()
```

## Cleaning Rate column (Replacing value like '4.1/5' value with value like 4.1)

```python
def cleanrate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
    
df1['rate']=df1['rate'].apply(cleanrate)
df1['rate'].head()
```

## Replacing NAN value with mean

```python
df1['rate'].fillna(df1['rate'].mean(),inplace=True)
df1['rate'].isnull().sum()
```

```python
df1.info()
```

## Drop Null values

```python
df1.dropna(inplace=True)
df1.head()
```

## Rename some columns

```python
df1.rename(columns={'approx_cost(for two people)':'platecost_for_2', 'listed_in(type)':'types', 'listed_in(city)':'city'}, inplace=True)
```

```python
df1.head()
```

```python
df1['location'].unique()
```

```python
df1['city'].unique()
```

## Drop City coulmn as It is same as location

```python
df1=df1.drop(['city'], axis=1)
```

```python
df1.head()
```

## cleaning types and platecost_for_2  columns 

```python
df1['types'].unique()
```

```python
def removecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df1['platecost_for_2']=df1['platecost_for_2'].apply(removecomma)
df1['platecost_for_2'].unique()
        
```

## Cleaning rest_type column

```python
rest_types=df1['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_test_type (value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df1['rest_type']=df1['rest_type'].apply(handle_test_type)
df1['rest_type'].value_counts()
```

## Cleaning location column

```python
lc=df1['location'].value_counts(ascending=False)
```

```python
lc_less=lc[lc<300]
lc_less
```

```python
def handle_location (value):
    if(value in lc_less):
        return 'others'
    else:
        return value
df1['location']=df1['location'].apply(handle_location)
df1['location'].value_counts()
```

```python
df1.head()
```

## Cleaning cuisines column

```python
cs=df1['cuisines'].value_counts()
```

```python
cs_less=cs[cs<100]
cs_less
def handle_cuisines (value):
    if(value in cs_less):
        return 'others'
    else:
        return value
df1['cuisines']=df1['cuisines'].apply(handle_cuisines)
df1['cuisines'].value_counts()
```

```python
df1.head()
```

```python
df1['types'].value_counts()
```
